```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'question_1.xlsx' is in the current directory or provide the correct path
file_path = '/content/drive/MyDrive/question_1.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')  # Load the DataFrame here

df['Total'] = df[['math score', 'reading score', 'writing score']].sum(axis=1) # Recalculate Total column

df['Total'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black', title='histogram of Total')
plt.xlabel('Total marks')
plt.ylabel('Frequency')
plt.show()
```
